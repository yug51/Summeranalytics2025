### 1. What is name of car that has the highest horsepower?  

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Cars.csv")

# Convert 'horsepower' to numeric, handle non-numeric values
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Drop rows with missing horsepower
df = df.dropna(subset=['horsepower'])

# Reset index (optional but helpful)
df = df.reset_index(drop=True)

# Find the name of the car with the highest horsepower
max_hp_row = df[df['horsepower'] == df['horsepower'].max()]
car_name = max_hp_row['name'].values[0]

print("Car with highest horsepower:", car_name)
```

### 2. How many cars have mpg  35?

```python
(df['mpg'] >= 35).sum()

```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Ensure numeric values
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

# Filter condition
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find most common origin
most_common_origin = filtered_df['origin'].mode()[0]
print("Most common origin:", most_common_origin)

```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
mean_acc_japan = round(df[df['origin'] == 'Japan']['acceleration'].mean(), 2)
print("Mean acceleration (Japan):", mean_acc_japan)
```

### 5. Which year had the highest average mpg?

```python
# Ensure 'mpg' is numeric
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# Group by year and calculate mean mpg
highest_avg_mpg_year = df.groupby('model_year')['mpg'].mean().idxmax()
print("Year with highest average mpg:", highest_avg_mpg_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Ensure numeric
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# Calculate median mpg
median_mpg = df['mpg'].median()

# Filter and calculate ratio
filtered = df[df['mpg'] > median_mpg].copy()
filtered['hp_weight_ratio'] = filtered['horsepower'] / filtered['weight']

# Get car(s) with max ratio
best_ratio_cars = filtered[filtered['hp_weight_ratio'] == filtered['hp_weight_ratio'].max()]
print(best_ratio_cars[['name', 'horsepower', 'weight', 'hp_weight_ratio']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by year and origin
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Clean numeric values
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# Create scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),
    alpha=0.7,
    hue_order=['japan', 'europe', 'usa']
)

plt.title('Horsepower vs Weight by Origin (Size by MPG)')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.show()

```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Ensure mpg is numeric
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# Group by car name
grouped = df.groupby('name')

# Create a summary DataFrame
summary = grouped.agg(
    count=('model_year', 'nunique'),  # number of distinct model years
    mpg_std=('mpg', 'std'),
    mpg_avg=('mpg', 'mean')
).reset_index()

# Filter for consistent cars
consistent_cars = summary[(summary['count'] >= 2) & (summary['mpg_std'] < 1.0)]

# Sort by count (desc), then avg mpg (desc)
consistent_cars_sorted = consistent_cars.sort_values(by=['count', 'mpg_avg'], ascending=[False, False])

# Display result
print(consistent_cars_sorted[['name', 'count', 'mpg_std', 'mpg_avg']])
```
